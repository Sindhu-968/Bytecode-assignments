Reverse a string without built-in reverse methods.

```python
s="Sindhu"
s1=s[::-1]
print(s1)
```

Check if a string is a palindrome.

```python
s="radar"
s2=s[::-1]
print(s2)
if s==s2:
  print("palindrome")
else:
  print("Not a palindrome")
```
