Write a Python program to find common elements in three lists using set?
Given lists:

l1 = [1, 5, 10, 20, 40, 80]

l2 = [6, 7, 20, 80, 100]

l3 = [3, 4, 15, 20, 30, 70, 80, 120]


```python
l1 = [1, 5, 10, 20, 40, 80]

l2 = [6, 7, 20, 80, 100]

l3 = [3, 4, 15, 20, 30, 70, 80, 120]
a=set(l1)
b=set(l2)
c=set(l3)
d=a.intersection(b,c)
print(d)
```

Write a Python program to find the longest word and smallest word in a set of words?
Words = {"apple", "banana", "cherry", "date"}


```python
words = {"apple", "banana", "cherry", "date"}

longest = max(words, key=len)
smallest = min(words, key=len)

print("Longest word:", longest)
print("Smallest word:", smallest)
```
